# River Flood Data from CLIMADA API

```python
from climada.util.api_client import Client

client = Client()
ISO = "FRA"
hazard = client.get_hazard("flood", properties={
    'res_meter': ['250'],
    'spatial_coverage': ['country'],
    'year_range': ['2002_2019'],
    'country_iso3alpha': ISO,
    'country_name': ['France']
    })
```

```python
hazard.plot_intensity(event=1)
```

## Impact Function

From Petals (but Manually)

https://climada-petals.readthedocs.io/en/stable/_modules/climada_petals/entity/impact_funcs/river_flood.html#ImpfRiverFlood.from_jrc_region_sector

```python
import numpy as np
from climada.entity import ImpactFunc

intensity = np.array([0., 0.05, 0.5, 1., 1.5, 2., 3., 4., 5., 6., 12.])
damage = [0., 0., 0.3, 0.55,  0.65, 0.75,  0.85, 0.95, 1., 1., 1.],

impf_FL = ImpactFunc(
    id=1,
    name = "Hail Impact Function",
    intensity_unit="mm",
    haz_type=hazard.haz_type,
    intensity=intensity,
    mdd=damage,
    paa = np.ones_like(intensity)
)
```

```python
impf_FL.plot()
```
