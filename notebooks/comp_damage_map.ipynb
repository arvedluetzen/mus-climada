```python
def comp_damage_map (eai, value, area):
    """Combines:
    Area per Departement
    Commulative EAI over different Hazards
    Relative Agriculture Area
    
    Into:
    Expected Annual Damaged Area
    """
    
    rel_damaged = eai * value
    damaged_area = area * rel_damaged
    
    return damaged_area
```

```python
from src.helpers import comp_impact

df = exposure.gdf
df["eai"] = comp_impact(haz_dict=haz_dict,
                        exposure=exposure)

df["damaged_area"] = comp_damage_map(
    eai=df["eai"],
    area=df["area"],
    value=df["value"]
)

df
```
