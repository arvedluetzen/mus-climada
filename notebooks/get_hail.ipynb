# Loading Hail Data

## Hazard Data from Climada API

```python

from climada.util.api_client import Client

client = Client()
ISO = "FRA"
hazard = client.get_hazard("hail", properties={"country_iso3alpha": ISO, 'climate_scenario': 'REF'})
```

```python
hazard.plot_intensity(event=0)
```

## Impact Function

```python
import numpy as np
from climada.entity import ImpactFunc

impf = ImpactFunc()
impf.haz_type = 'HL' # hail
impf.id = 1
impf.name = 'Hail impact on crops (diameter-based)'

# Intensity = hail diameter in mm
impf.intensity = np.array([0, 6, 10, 20, 30, 50])

# Mean Damage Degree (0–1)
impf.mdd = np.array([0.0, 0.05, 0.15, 0.35, 0.60, 0.90])

# Percent of assets affected (assume all exposed crops affected)
impf.paa = np.ones_like(impf.intensity)

impf.check()
```

```python
impf.plot()
```
